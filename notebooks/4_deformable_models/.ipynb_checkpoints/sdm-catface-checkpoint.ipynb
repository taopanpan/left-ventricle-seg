```python
from pathlib import Path

import menpo.io as mio
```

## Recover Training Data

```python
import json

filename = 'data/tfrecords/lv/LV_data_manifest.json'

print('Reading JSON from {}'.format(filename))
with open(filename, 'r') as handle:
    dic = json.load(handle)
```

# Supervised Descent Method

### Preprocess Images for Training

```python
def process(image, crop_proportion=0.2, max_diagonal=400):
    if image.n_channels == 3:
        image = image.as_greyscale()
    image = image.crop_to_landmarks_proportion(crop_proportion)
    d = image.diagonal()
    if d > max_diagonal:
        image = image.rescale(float(max_diagonal) / d)
    return image

def sort_images(images):    
    train_indices = []
    eval_indices = []
    test_indices = []
    
    for i in range(len(images)):
        
        image = images[i]
    
        if str(image.path) in dic['train']['paths']:
            train_indices.append(i)
        elif str(image.path) in dic['validation']['paths']:
            eval_indices.append(i)
        elif str(image.path) in dic['test']['paths']:
            test_indices.append(i)
            
    train_images = images[train_indices]
    eval_images = images[eval_indices]
    test_images = images[test_indices]
    
    return train_images, eval_images, test_images

path_to_images = 'D:\dev\project\data\ibug\cats\images'
images = mio.import_images(path_to_images, verbose=True)

train_images, eval_images, test_images = sort_images(images)
```

```python
%matplotlib inline
from menpowidgets import visualize_images
visualize_images(train_images)
```

### Train the AAM

```python
from menpofit.sdm import SupervisedDescentFitter, NonParametricNewton

sdm_fitter = SupervisedDescentFitter(train_images, group='LJSON',
                                 sd_algorithm_cls=NonParametricNewton,
                                 verbose=True)
```

## Fit the AAM on new images

```python
from pathlib import Path

# LOAD INFERENCE IMAGES

path_to_images = Path('NOOBS.pkl')
images = mio.import_pickle(path_to_images)

visualize_images(images)
```
